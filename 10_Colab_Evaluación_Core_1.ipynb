# **Ejecución de la Actividad**

## **Misión 1: El Top 3 de México (Extracción, Filtro Múltiple y Límite)**

```python
%%sql

select pais,categoria,monto_usd
from ventas_globales
where pais = 'México'
and categoria = 'Software'
order by monto_usd desc
limit 3;
```

## **Misión 2: El Balance Global (Agregación sin filtros)**

```python
%%sql

select
SUM(monto_usd)AS Ingreso_Global,
AVG(satisfaccion_cliente)AS Satisfaccion_Promedio
FROM ventas_globales;
```

## **Misión 3: Análisis de Riesgo en Hardware (Agregación + Filtro)**

```python
%%sql

SELECT
COUNT(*) AS Total_Transacciones,
MIN(monto_usd) AS Peor_Venta,
MAX(monto_usd) AS Mejor_Venta
FROM ventas_globales
WHERE pais = 'Chile'
AND categoria = 'Hardware';
```
