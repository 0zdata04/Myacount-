````python
import pandas as pd

df_clientes = pd.read_csv("/content/20_CORE_clientes.csv")
df_transacciones = pd.read_csv("/content/19_CORE_transacciones.csv")
````

 **Misión 1: Estandarización de Variables (Clean Code)**

Antes de cruzar o filtrar, debes arreglar el desastre tipográfico en la columna ciudad de la tabla de clientes.

````python
#Normalizar la columna ciudad para asegurar que registros equivalentes
#(por ejemplo, "Santiago", "SANTIAGO" y "santiago ") sean tratados como un mismo valor.

df_clientes['ciudad'] = df_clientes["ciudad"].str.lower().str.strip()

# Validar el resultado de la limpieza revisando los valores únicos.
df_clientes["ciudad"].unique()
````

**Misión 2: Tratamiento de Outliers (Filtros Lógicos)**

Debes limpiar las anomalías matemáticas de ambas tablas antes de unirlas, justificando tu criterio.

**Depuración de clientes:**
Se filtrarán los registros para conservar únicamente clientes cuya edad se encuentre dentro del rango permitido por la empresa (18 a 99 años), eliminando posibles errores de captura o registros que no cumplan los requisitos establecidos.

````python
df_clientes = df_clientes[
(df_clientes["edad"]>= 18)&
(df_clientes["edad"]< 100)
]
````

**Validación de ingresos comerciales:**
Con el fin de asegurar la calidad del reporte comercial, se considerarán únicamente transacciones con montos mayores a cero. De esta forma se excluyen registros que podrían corresponder a errores operacionales, anulaciones o devoluciones, evitando que afecten los indicadores de ingresos.

````python
df_transacciones = df_transacciones[
(df_transacciones["monto_usd"] > 0)
]
````

 **Misión 3: Integración de Fuentes (Merge)**

**Decisión de unión de datos:**
Para este análisis se utilizará un INNER JOIN, ya que el objetivo es evaluar los ingresos generados por ciudad. Este tipo de unión permite conservar únicamente los clientes que poseen transacciones asociadas, excluyendo aquellos que no registran compras. De esta manera, el análisis se enfoca exclusivamente en los clientes que generaron ingresos reales, obteniendo resultados más representativos para el reporte comercial.

````python
df_final = pd.merge(
df_clientes,
df_transacciones,
on="id_cliente",
how ="inner"
)
````

**Misión 4: Gestión Final de Nulos (Sanity Check)**

````python
df_final.isna().sum()
````

**(Justificación):** Se decidió reemplazar los valores nulos por la etiqueta **Sin Registro** en lugar de eliminar las filas correspondientes. Esta decisión permite conservar la información de las transacciones y evitar la pérdida de ingresos válidos dentro del análisis. Si se eliminaran estas filas, también se eliminarían ventas reales, lo que podría generar resultados incorrectos y afectar la precisión de los indicadores comerciales. La etiqueta "Sin Registro" permite identificar claramente los datos faltantes sin comprometer la integridad de la información disponible.

````python
df_final['metodo_pago'] = df_final['metodo_pago'].fillna("Sin Registro")
````
