##5.**Construcion del Dataset Maestro**

```python
%%sql
SELECT
-- Información temporal
p.Fecha_Pedido,
p.ID_Pedido,

-- Información cliente
c.ID_Cliente,
c.Nombre_Cliente,
c.Segmento,

-- Información geográfica
g.ciudad,
g.region,

-- Información producto
pr.ID_producto,
pr.Categoria,
pr.Subcategoria,
pr.Nombre_Producto,

-- Métricas comerciales
dp.Ventas,
dp.Ganancia,
dp.Cantidad,
dp.Descuento

FROM "Detalle_Pedido" AS dp

-- INNER JOIN 1
INNER JOIN Pedidos as p
  ON dp.ID_Pedido = p.ID_Pedido

-- INNER JOIN 2
INNER JOIN Clientes as c
  ON p.ID_Cliente = c.ID_Cliente

--INNER JOIN 3
INNER JOIN Cliente_Ubicacion as  cu
  ON c.ID_Cliente = cu.ID_Cliente

-- INNER JOIN 4
INNER JOIN Geografia as g
  ON cu.ID_Ubicacion = g.ID_Ubicacion

-- INNER JOIN 5
INNER JOIN Productos as  pr
  ON  dp.ID_Producto = pr.ID_Producto
```

##8.**Creación de la primera métrica calculada-(CASE WHEN)**

```python
%%sql

SELECT

c.ID_Cliente,

c.Nombre_Cliente,

SUM(dp.Ventas) AS Total_Ventas,

CASE

WHEN SUM(dp.Ventas) >= 10000 THEN 'Cliente VIP'

WHEN SUM(dp.Ventas) >= 5000  THEN 'Cliente Frecuente'

ELSE 'Cliente Ocasional'

END AS Clasificacion_Clientes

FROM Detalle_Pedido dp

INNER JOIN Pedidos  p
ON dp.ID_Pedido = p.ID_Pedido

INNER JOIN Clientes c
ON p.ID_Cliente = c.ID_Cliente

GROUP BY

c.ID_Cliente,
c.Nombre_Cliente;
```

**Justificación de negocio**

Se creó una variable calculada utilizando CASE WHEN para clasificar a los clientes según el total de ventas generadas. Esta métrica permite identificar distintos niveles de aporte económico al negocio y facilita la definición de estrategias comerciales orientadas a la fidelización y retención de clientes.

---
**Justificación de negocio**

Se relaciona la tabla **Detalle_Pedido** con **Pedidos** para acceder a la información del cliente asociada a cada transacción. Esta relación permite conectar las ventas registradas con el cliente que realizó la compra, siendo un paso indispensable para calcular el total de ventas por cliente.

**Justificación de negocio**

Se relaciona la tabla **Pedidos** con **Clientes** para identificar a cada comprador mediante su nombre e identificador. Esta integración permite agrupar las ventas por cliente y construir una clasificación basada en su aporte económico al negocio.
