# **Ejecución de la Actividad**

```python
%%sql
/*
    Reporte: Ingreso Mensual Recurrente por Cliente Corporativo
    Solicitado por: Director Financiero (CFO) — TechNova SaaS
    Objetivo: Calcular el total de pagos mensuales activos por cada cliente
    corporativo, excluyendo cuentas canceladas o atrasadas.
*/
SELECT
u.nombre_completo AS Cliente,
SUM(s.pago_mensual_usd) AS Ingreso_mensual
FROM usuarios AS u
INNER JOIN suscripciones AS s
ON u.id_usuario = s.id_usuario
-- Solo proyección de ingresos seguros: Clientes Al día y de tipo Corporativo
WHERE u.tipo_cliente = 'Corporativo' AND s.estado_cuenta = 'Al día'
GROUP BY u.id_usuario
ORDER BY Ingreso_mensual DESC;
-- Lee esta consulta y entiende por qué funciona.
```
